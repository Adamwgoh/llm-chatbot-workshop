# App Deployment

In today's main course, you need to combine all you have learned in the previous 3 chapters and modify this repository such that:

- It is able to support more than 2 non ollama models. Make sure to embed your query to the right embedding!
- Augment your chatbot with a in-memory vector database. For extra challenge, host a vectorDB and connect to it
- Deploy your chatbot to deployment platform of your choice
    - vercel
    - fly.io
    - render

- Modify your search to make it more accurate

Bonus
- Support multi-model GPT-4o
